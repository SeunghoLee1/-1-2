```python
import time
import operator
from konlpy.tag import Okt #Okt 라이브러리 부르기(Window에선 가장 빠르다)
start_time = time.time()
f = open('어린왕자.txt',encoding = 'utf8')#파일 열기
cnt = {}
list = []
okt = Okt()
for row in f:
    k = row.strip()#처리 위해 필요
    list.append(okt.pos(k))#list에 그 단어와 품사로 이루어진 튜플을 계속 추가
print("--- %s seconds ---" %(time.time() - start_time))
for i in list:
    for j in i:
        if j[1] in cnt:#품사가 그 안에 있으면
            cnt[j[1]] += 1#그 항목의 key값 1 더하기
        else:#아니면
            cnt[j[1]] = 1#새로운 항목 만들기
sortedArr2 = sorted(cnt.items(), key=operator.itemgetter(1), reverse = True)#갯수가 많은 순으로 정렬
print(sortedArr2)
print("--- %s seconds ---" %(time.time() - start_time))
```
