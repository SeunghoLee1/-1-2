```python
import matplotlib.pyplot as plt
import operator
f = input('어떤 파일을 사용할 것인가요? 꼭 Download 폴더에 넣으시고 확장명은 txt로')#사용할 파일명 입력
sum = 0 #모든 글자 갯수
cnt = {} #갯수를 셀 딕셔너리
y = open(f)
alpha = ['a', 'b', 'c','d', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
#모든 알파벳(소문자인 이유는 lower()함수때문)
for row in y :
    k = row.strip()#한줄씩 나눠서
    k = k.lower()#대소문자통합
    for i in k:
        if i in alpha: #\n 등을 빼기 위해
            if i in cnt:
                cnt[i] = cnt[i] + 1
            else:
                cnt[i] = 1
            # 갯수세기
for i in cnt:
    sum = sum + cnt[i]# 모든 글자 갯수 합치기
sortedArr1 = sorted(cnt.items())#알파벳순으로 정렬
sortedArr2 = sorted(cnt.items(), key=operator.itemgetter(1), reverse = True)#갯수가 많은 순으로 정렬
for i in cnt:
    print(cnt[i]*100/sum, i)# 각 글자의 비율을 %로
print(sortedArr1)
print(sortedArr2)


plt.bar(range(len(cnt.keys())),cnt.values())
plt.xticks(range(len(cnt.keys())),cnt.keys(), rotation = 90)
plt.show() #그래프 표현
```
