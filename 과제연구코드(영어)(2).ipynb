```python
import matplotlib.pyplot as plt
import operator
f = input('   ?  Download    txt')#  
sum = 0 #  
cnt = {} #  
y = open(f)
alpha = ['a', 'b', 'c','d', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
# (  lower())
for row in y :
    k = row.strip()# 
    k = k.lower()#
    for i in k:
        if i in alpha: #\n   
            if i in cnt:
                cnt[i] = cnt[i] + 1
            else:
                cnt[i] = 1
            # 
for i in cnt:
    sum = sum + cnt[i]#    
sortedArr1 = sorted(cnt.items())# 
sortedArr2 = sorted(cnt.items(), key=operator.itemgetter(1), reverse = True)#   
for i in cnt:
    print(cnt[i]*100/sum, i)#    %
print(sortedArr1)
print(sortedArr2)


plt.bar(range(len(cnt.keys())),cnt.values())
plt.xticks(range(len(cnt.keys())),cnt.keys(), rotation = 90)
plt.show() # 
```
