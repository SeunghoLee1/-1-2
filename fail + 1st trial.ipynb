```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
import tensorflow as tf#seed 값 생성
import matplotlib.pyplot as plt
seed = 0
numpy.random.seed(seed)
tf.set_random_seed(seed)

#data 로드
dataset = numpy.loadtxt("데이터(기사).csv", encoding = 'cp949', delimiter=",")
X = dataset[:,0:21]
Y = dataset[:,21]

#data 전처리

#모델 설정
model = Sequential()
model.add(Dense(12, input_dim=6, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

#모델 컴파일
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

#모델 실행
model.fit(X, Y, epochs=200, batch_size=10)

#결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
import tensorflow as tf#seed 값 생성
import matplotlib.pyplot as plt
seed = 0
numpy.random.seed(seed)
tf.set_random_seed(seed)

#data 로드
dataset = numpy.loadtxt("데이터(기사)-단어없음.csv", encoding = 'cp949', delimiter=",")
X = dataset[:,0:6]
Y = dataset[:,6]

#data 전처리

#모델 설정
model = Sequential()
model.add(Dense(12, input_dim=6, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

#모델 컴파일
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

#모델 실행
hist = model.fit(X, Y, epochs=200, batch_size=10)

#결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))

fig, acc_ax = plt.subplots()
acc_ax.plot(hist.history['acc'], 'b', label='accuracy')
acc_ax.set_ylabel('accuracy')
acc_ax.legend(loc='upper left')

plt.show()
```
