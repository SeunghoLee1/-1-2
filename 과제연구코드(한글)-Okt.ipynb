```python
import time
import operator
from konlpy.tag import Okt
start_time = time.time()
f = open('어린왕자.txt',encoding = 'utf8')
cnt = {}
list = []
okt = Okt()
for row in f:
    k = row.strip()
    list.append(okt.pos(k))
print("--- %s seconds ---" %(time.time() - start_time))
for i in list:
    for j in i:
        if j[1] in cnt:
            cnt[j[1]] += 1
        else:
            cnt[j[1]] = 1
sortedArr2 = sorted(cnt.items(), key=operator.itemgetter(1), reverse = True)#갯수가 많은 순으로 정렬
print(sortedArr2)
print("--- %s seconds ---" %(time.time() - start_time))
```
