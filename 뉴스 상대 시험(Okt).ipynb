```python
import operator
from konlpy.tag import Okt #Okt 라이브러리 부르기(Window에선 가장 빠르다)
f = open('어린왕자.txt',encoding = 'utf8')#파일 열기
cnt = {}
list = []
okt = Okt()
for row in f:
    k = row.strip()#처리 위해 필요
    list.append(okt.pos(k))#list에 그 단어와 품사로 이루어진 튜플을 계속 추가
for i in list:
    for j in i:
        if j[1] in cnt:#품사가 그 안에 있으면
            cnt[j[1]] += 1#그 항목의 key값 1 더하기
        else:#아니면
            cnt[j[1]] = 1#새로운 항목 만들기
print(cnt)
print(cnt['Noun'])
print(cnt['Adjective'])
print(cnt['Verb'])
print(cnt['Adverb'])
print(cnt['Conjunction'])
print(cnt['VerbPrefix'])
noun = {}
adjective = {}
verb = {}
adverb = {}
josa = {}
for i in list:
    for j in i:
        if j[1] == 'Noun':
            if j[0] in noun:
                noun[j[0]] += 1
            else:
                noun[j[0]] = 1
sortedArrnoun = sorted(noun.items(), key=operator.itemgetter(1), reverse = True)#갯수가 많은 순으로 정렬
print(sortedArrnoun[0:3])
for i in list:
    for j in i:
        if j[1] == 'Adjective':
            if j[0] in adjective:
                adjective[j[0]] += 1
            else:
                adjective[j[0]] = 1
sortedArradj = sorted(adjective.items(), key=operator.itemgetter(1), reverse = True)#갯수가 많은 순으로 정렬
print(sortedArradj[0:3])
for i in list:
    for j in i:
        if j[1] == 'Verb':
            if j[0] in verb:
                verb[j[0]] += 1
            else:
                verb[j[0]] = 1
sortedArrverb = sorted(verb.items(), key=operator.itemgetter(1), reverse = True)#갯수가 많은 순으로 정렬
print(sortedArrverb[0:3])
for i in list:
    for j in i:
        if j[1] == 'Adverb':
            if j[0] in adverb:
                adverb[j[0]] += 1
            else:
                adverb[j[0]] = 1
sortedArradverb = sorted(adverb.items(), key=operator.itemgetter(1), reverse = True)#갯수가 많은 순으로 정렬
print(sortedArradverb[0:3])
for i in list:
    for j in i:
        if j[1] == 'Josa':
            if j[0] in josa:
                josa[j[0]] += 1
            else:
                josa[j[0]] = 1
sortedArrjosa = sorted(josa.items(), key=operator.itemgetter(1), reverse = True)#갯수가 많은 순으로 정렬
print(sortedArrjosa[0:3])
```
